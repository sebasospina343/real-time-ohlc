```python
import os
from pathlib import Path

# Function to load .env files with export format
def load_env_file(env_path):
    """Load environment variables from a file that uses 'export KEY=VALUE' format"""
    env_path = Path(env_path)
    if env_path.exists():
        with open(env_path, 'r') as f:
            for line in f:
                line = line.strip()
                if line and not line.startswith('#') and line.startswith('export '):
                    # Remove 'export ' prefix and parse KEY=VALUE
                    key_value = line[7:].strip()  # Remove 'export '
                    if '=' in key_value:
                        key, value = key_value.split('=', 1)
                        # Remove quotes if present
                        value = value.strip('"\'')
                        os.environ[key.strip()] = value.strip()
                        print(f"Loaded: {key.strip()}")

# Load .env files relative to notebook location
# The notebook is in services/price_predictor/noteboooks/
# The .env files are in services/price_predictor/
current_dir = Path.cwd()
if current_dir.name == 'noteboooks':
    parent_dir = current_dir.parent
else:
    # Fallback: try relative path
    parent_dir = Path('../services/price_predictor') if not (current_dir / '.live.env').exists() else current_dir

load_env_file(parent_dir / '.live.env')
load_env_file(parent_dir / '.tools2.env')
```
